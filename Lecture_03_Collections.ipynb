```python
l1=[1,2,3,4]
l2=l1         #<- you are not actually copying the list. You just create a reference to the same
print(l2)
l1[0]=100
print()
print("l1",l1)
print("l2",l2)
# we are not creating a new list, but we are modifying the same list that l1 and l2 are referencing (same entity with two different names)
# if you want to create a new list, you can use the copy method
# l2=l1.copy()  #<- this is how you create a new list that is a copy of l1
# or using the list comprehension or using the list constructor
```

```python
l1
```

```python
#Using the constructor of lists (one method to create a copy of a list)
l2=list(l1) #l2 is a copy and l1 changes the first element
l1[0]=1
print(l1)
print(l2)
```

```python
l1
```

```python
#Using list comprehensions
l2=[ x for x in l1 ] # [1, 2, 3, 4] 
l1[0]=300
print(l1)
print(l2)
```

```python
[x*2 for x in l1]
```

```python
[x**2 for x in l1] # we do the power of 2 to each element of the list l1
```

```python
l3=l2.copy()
l2[0]=55
print(l2)
print(l3)
```

Tuples: an immutable version of lists

Tuples are another sequence type.

- They are essentially immutable lists
- They are created using rounded parenthesis rather than squared ones (or not parenthesis at all)

```python
lst=['a','b','c']
key=1
print('Integer',key,'     is mapped to char',lst[key])

marks={ 'Andrea':30, 
        'Daniele':29, 
        'Giulio':25 }
print('Student','Andrea','is mapped to mark',marks['Andrea'])
```
