Statement
Write a program that

- Reads a string s
- Creates a list lst containing as elements the characters in s in the same order in which they appear in s
    - sets the string '1' as the value of the first entry of lst
    - prints lst
- Creates a set st containing the characters in the list
    - prints the length of st
    - prints the minimum element in st
    - prints the maximum element in st
- Creates a list lst2 containing the elements in st
    - prints the length of lst2
    - prints the minimum element in lst2
    - prints the maximum element in lst2
    - sorts lst2
    - appends string '!' at the end of lst2
    - prints lst2


```python
# STEP 1:
s = input()
print(s)
```

```python
# STEP 2:
lst = list(s)
lst[0]='1'
print(lst)
```

```python
# STEP 3:
st = set(lst)
print(len(st))
print(min(st))
print(max(st))
```

```python
# STEP 4:
lst2 = list(st)
print(len(lst2))
print(min(lst2))
print(max(lst2))
lst2.sort() # ordina la lista in ordine crescente
lst2.append('!') # si può fare anche lst2[-1]= '!' ma append è più semplice
print(lst2)
```

```python
# STEP 1 TO 4:
s = input()

lst = list(s)
lst[0] = '1'
print(lst)

st = set(lst)
print(len(st))
print(min(st))
print(max(st))

lst2 = list(st)
print(len(lst2))
print(min(lst2))
print(max(lst2))

lst2.sort()
lst2.append('!')
print(lst2)
```
