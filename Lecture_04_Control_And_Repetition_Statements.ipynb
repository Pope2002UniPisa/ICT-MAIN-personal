If statement:

    This section presents the various forms in which we can use the if statement

    Choose whether executing a block or not
    In its simplest form, an if statement consists of

- A guard: a Boolean expression (as presented in class 2 evaluated to either True or False)
- A body: block of statements executed only if the guards evaluates to True

```python
x=1
y=2
if x==y:
    print('The first guard was satisfied')
    print('ciao')
# In Python quando si esce dal ciclo if, si esce anche dal blocco di codice indentato, quindi se x!=y, non viene eseguito il print e si passa alla riga successiva
print('After the first if\n')
# \n fa lasciare una riga vuota dopo il print
x=1
y=1
if x==y :
    print('\tThe second guard was satisfied')
    #\t fa lasciare uno spazio prima del print
    print('ciao')
print('After the second if')
```
